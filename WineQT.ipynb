Calculate summary statistics (mean, median, standard deviation) for a chosen feature.

```python
chosen_feature = 'alcohol'
mean_val = df[chosen_feature].mean()
median_val = df[chosen_feature].median()
std_dev = df[chosen_feature].std()

print(f"Summary Statistics for '{chosen_feature}':")
print(f"Mean: {mean_val:.2f}")
print(f"Median: {median_val:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
```

```python
# Define key features
key_features = ['alcohol', 'volatile acidity', 'citric acid', 'residual sugar', 
                'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 
                'density', 'pH', 'sulphates']

# Plot distribution of key features
plt.figure(figsize=(15, 10))
for i, feature in enumerate(key_features, 1):
    plt.subplot(3, 4, i)
    sns.histplot(df[feature], kde=True)
    plt.title(feature)
plt.tight_layout()
plt.show()

# Compare features across different quality ratings
plt.figure(figsize=(15, 10))
for i, feature in enumerate(key_features, 1):
    plt.subplot(3, 4, i)
    sns.boxplot(x='quality', y=feature, data=df)
    plt.title(feature)
plt.tight_layout()
plt.show()
```
